```python
import nibabel as nib
import matplotlib.pyplot as plt
# Load and visualize additional NIfTI files
for run in [2, 3]:
    img = nib.load(f'/home/lassub/TTK/Lifespan/AOMIC_external_data/ds003097/sub-0001/anat/sub-0001_run-{run}_T1w.nii.gz')
    data = img.get_fdata()
    slice_index = data.shape[2] // 2
    plt.imshow(data[:, :, slice_index], cmap='gray')
    plt.title(f"Run {run} - Slice {slice_index}")
    plt.axis('off')
    plt.show()
# Load the NIfTI file
img = nib.load('/home/lassub/TTK/Lifespan/AOMIC_external_data/ds003097/sub-0001/anat/sub-0001_run-1_T1w.nii.gz')
data = img.get_fdata()

# Print the available number of slices
print(f"Available slices: {data.shape[2]}")
# Visualize a slice (e.g., the middle slice)
slice_index = data.shape[2] // 2
plt.imshow(data[:, :, slice_index], cmap='gray')
plt.title(f"Slice {slice_index}")
plt.axis('off')
plt.show()
```
